Q2(a)

Create a function to solve ax+by=c
Divide by gcd of a and b
Iterate Euclid's algorithm and then multiply back by a factor of c
Then substitute values for the equation given

```python
import math

def euc(a,b,c):

    d = math.gcd(a,b)
    if c%d!=0:
        raise ValueError("No Solutions")
    a=a//d
    b=b//d
    c=c//d
    r0,r1=a,b
    x0,x1=1,0
    y0,y1=0,1
    while r1!=0:
        x0,x1=x1,x0-(r0//r1)*x1
        y0,y1=y1,y0-(r0//r1)*y1
        r0,r1=r1,r0-(r0//r1)*r1
    x=x0*c
    y=y0*c
    return x,y
    
#solve 993319x+999331y=939193

print(euc(993319,999331,939193))    
```
