Q3

Plot each sequence with a logarithmic scale on the y-axis

```python
import matplotlib.pyplot as plt

A = list(range(1, 101))
a = [2**(-n) for n in A]

plt.figure()
plt.semilogy(A,a)
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("(a)")
plt.legend()
plt.show()

B = list(range(1, 101))
b = [1/(n + 1) for n in B]

plt.figure()
plt.semilogy(B,b)
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("(b)")
plt.legend()
plt.show()

C = list(range(1, 11))
c = [2**(-(2**n)) for n in C]

plt.figure()
plt.semilogy(C,c)
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("(c)")
plt.legend()
plt.show() 
```

Q5 (c)

Create lists of a_n, Aitken acceleration and iterated Aitken acceleration partial sums using formula given up to n=10000
Create lists for their errors
Plot each sequence of errors


```python
import mpmath
import matplotlib.pyplot as plt
mpmath.mp.dps = 100

a=[0]
for i in range(10002):   
    a.append(a[-1]+4*(-1)**i/(2*i + 1))

accel=[]
for i in range(10000): 
    accel.append(a[i]-(a[i+1]-a[i])**2/(a[i+2]-2*a[i+1]+a[i]))

iter_accel=[]
for i in range(9997): 
    iter_accel.append(accel[i]-(accel[i+1]-accel[i])**2/(accel[i+2]-2*accel[i+1]+accel[i]))

e_a=[]
for i in range(10002):
    e_a.append(abs(a[i]-mpmath.pi))

e_accel=[]
for i in range(9999):
    e_accel.append(abs(accel[i]-mpmath.pi))    

e_iter_accel=[]
for i in range(9997):
    e_iter_accel.append(abs(iter_accel[i]-mpmath.pi))    

plt.figure()
plt.semilogy(e_a,label="Partial sums")
plt.semilogy(e_accel, label="Aitken acceleration")
plt.semilogy(e_iter_accel,label="Iterated Aitken acceleration")
plt.xlabel("nth term")
plt.ylabel("error")
plt.legend()
plt.show()     
```
