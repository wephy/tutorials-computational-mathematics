Q3 (a)

Create iteration using Newton's formula and use the given x0=1/2. 
Use function given in question and it's derivative for each part.
Stop iteration when function is sufficiently close to 0 as specified in question.
Find order of convergence for each iteration and then average across all iterations.

```python
import math

x=0.5
values=[x]
orders=[]

def f(x):
    return math.sin(x)*math.sinh(x)
def f_prime(x):
    return math.cos(x) * math.sinh(x) + math.sin(x) * math.cosh(x)

while abs(f(x))>1e-8:
    x=x-f(x)/f_prime(x)
    values.append(abs(x))

for n in range(2, len(values)-1):
    p=(math.log(values[n+1]/values[n]))/(math.log(values[n]/values[n-1]))
    orders.append((p))

q=sum(orders)/len(orders)

print(f"root is approximately {x}")
print(f"Has {q} order of convergence")
```

f'(0)=0, f''(0)≠0 so observed linear order of convergence matches expected linearity form Q2(b)
