The question posed is to find integer solutions x,y such that ax + by = c for integers a,b,c if any such solutions exist.
First we check if any solutions exist by ensuring c is divisible by the GCD of a and b, otherwise the left hand side divides GCD(a,b) but the right hand side does not

```python
import math

#want solutions to ax + by = c if any exist

a = int(input("Enter a:"))
a1 = a
b = int(input("Enter b:"))
b1 = b
c = int(input("Enter c:"))
c1 = c

#ensuring a greater
if b > a:
    temp = b
    b = a
    a = temp

while a > 1 and b > 1:
    d = a//b
    if a%b == 0:
        break
    temp = b
    b = a - d*b
    a = temp
print("GCD of a,b is: ",b)
if c%b != 0:
    print("no integer solutions")

```

Now, if solutions do exist, we can solve for ax + by = GCD(a,b), and then multiply x,y by c/(GCD) to get solutions to ax + by = c.

```python
def solution(a, b):
    if a == 0:
        return b, 0, 1
    gcd, x1, y1 = solution(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return gcd, x, y

if c%b == 0:
    multiplier = c/b #this is the factor needed since the algorithm returns ax + by = GCD(a,b)
    a = a1
    b = b1
    c = c1
    sol = solution(a,b)
    x = sol[1]
    y = sol[2]
    x = x*multiplier
    y = y*multiplier
    print("x=",x)
    print("y=",y) 
    print (a,"*",x,"+",b,"*",y,"=",c)
```

The next question involves RSA. L(n) = (p-1)(q-1) / GCD(p-1,q-1)
The associated private key satisfies d*e = 1 mod L(n)
Using the previous part, we want the solution to ex + L(n)y = 1, as then d = x

```python
p = int(input("Enter p:"))
q = int(input("Enter q:"))
e = int(input("Enter e"))
temp = solution(p,q)
Ln = (p-1)*(q-1) / (temp[0])
print(Ln)
sol = solution(e, Ln)
d = sol[1]
while d < 0:
    d += Ln
print(d)

```
