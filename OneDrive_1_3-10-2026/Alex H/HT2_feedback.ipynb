Classify the order of convergence for each of the following sequences:
a) 2^(-n) order of convergence = 1 (linear)
b) 1/(n+1) order of convergece is sublinear
c) 2^ (- 2^(n) ) order of convergence is 2 (quadratic/superinear)
For each sequence plot a(n) from n = 1 to N, N = 100 for a,b  and N = 10 for c

Question 4:  Does Banach’s contraction mapping theorem apply to the following continuous real-valued
functions g with domain of definition X?
a) X = [3pi/4 , 5pi/4]  g(x) = sin(x) + x
Yes, as g(x) in X and g'(x) = 1 + cosx, which is between -1 and 1 for x in X
b) X = [0,2]  g(x) = x^3 + 3x^2 -3
No, g(0) = -3, -3 not in X so not self mapping
c) X = [1, infinty)  g(x) = x + 1/x
No, even though g(x) >= x, so g(x) is self-mapping, g'(x) = 1 -  x^-2, and sup(1 - x^-2) over [1, infinity) = 1, which is not strictly less than 1, so is not a contraction.
