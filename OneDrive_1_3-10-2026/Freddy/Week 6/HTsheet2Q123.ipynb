```python
import numpy as np

x=0.5

while abs(np.sin(x)*np.sinh(x))>1e-8:
    x=x-((np.sin(x)*np.sinh(x))/(np.sin(x)*np.cosh(x)+np.cos(x)*np.sinh(x)))

print(x)
```

f'(x)=0 and f''(x) does not =0 so convergence is linear, contracting by 1/2

```python
import numpy as np

x=0.5

while abs(x-np.sin(x))>1e-8:
    x=x-((x-np.sin(x))/(1-np.cos(x)))

print(x)
```

f'(x)=f''(x)=0, f'''(x) does not =0, so convergence is linear, contraction 2/3

```python
import numpy as np

x=0.5

while abs(x*np.cos(x))>1e-8:
    x=x-((x*np.cos(x))/(np.cos(x)-x*np.sin(x)))

print(x)
```

f''(x)=0 but f'(x) does not =0, so convergence is cubic
