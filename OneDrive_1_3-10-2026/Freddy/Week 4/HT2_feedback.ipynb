```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linspace

def f(n):
    return 2**(-n)

xsa = linspace(1,100,100)
ysa = np.log(f(xsa))

plt.plot(xsa,ysa)
plt.show()


def g(n):
    return (n+1)**(-1)

xsb = linspace(1,100,100)
ysb = np.log(g(xsb))

plt.plot(xsb,ysb)
plt.show()



def h(n):
    return 2**((-2**n))

xsc = linspace(1,10,10)
ysc = np.log(h(xsc))

plt.plot(xsc,ysc)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linspace

def A(n):
    n=int(n)
    return 4*sum(((-1)**k)/(2*k+1) for k in range(0,n+1))

def E(n):
    n=int(n)
    return abs(A(n)-np.pi)

def A1(n):
    n=int(n)
    return A(n)-((A(n+1)-A(n))**2)/(A(n+2)-2*A(n+1)+A(n))

def E1(n):
    n=int(n)
    return abs(A1(n)-np.pi)

def A2(n):
    n=int(n)
    return A1(n)-((A1(n+1)-A1(n))**2)/(A1(n+2)-2*A1(n+1)+A1(n))

def E2(n):
    n=int(n)
    return abs(A2(n)-np.pi)

xs = linspace(1,10000, 500)
ys = np.array([E(x) for x in xs])
ys1 = np.array([E1(x) for x in xs])
ys2 = np.array([E2(x) for x in xs])

plt.plot(xs,ys)
plt.plot(xs,ys1)
plt.plot(xs,ys2)

plt.yscale("log")

plt.show()
```
