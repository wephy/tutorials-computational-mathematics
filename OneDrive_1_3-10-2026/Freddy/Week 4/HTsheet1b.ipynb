```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linspace

def f(n):
    return 2**(-n)

xsa = linspace(1,100,100)
ysa = np.log(f(xsa))

plt.plot(xsa,ysa)
plt.show()


def g(n):
    return (n+1)**(-1)

xsb = linspace(1,100,100)
ysb = np.log(g(xsb))

plt.plot(xsb,ysb)
plt.show()



def h(n):
    return 2**((-2**n))

xsc = linspace(1,10,10)
ysc = np.log(h(xsc))

plt.plot(xsc,ysc)
plt.show()


```

```python
import numpy as np
import mpmath
mpmath.mp.dps = 100

def f(n):
    return abs(4*sum(((-1)**k)/(2*k+1) for k in range(0,n+1))-np.pi)

En = f(10000)
En1 = f(10001)
En2 = f(10002)

q = (np.log(En2)-np.log(En1))/(np.log(En1)-np.log(En))

M = np.exp(((np.log(En1))**2-np.log(En2)*np.log(En))/(np.log(En1)-np.log(En)))

print(f"q = {q}")
print(f"M = {M}")
```

Hence the order of convergence is 1
