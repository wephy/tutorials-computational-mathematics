## Question 1

First note the base case. When $N=1$ the hypothesis requires $a\geq 2, b\geq 1$, but naturally as $a>b>0$ this is vacuous. Naturally the algorithm  requires at least 1 division when its inputs are positive. Thus the statement holds for $N\leq 1$.

Suppose the statement holds for all $N \leq M$.
Then consider a pair $a>b$ which requires at least $M+1$ divisions in the algorithm. Thus the pair requires at least $M$ divisions; hence by inductive hypothesis $a\geq F_{M+2}, b\geq F_{M+1}$.
The first division yields $a=bq_0+r_0$ and thus provides a new seed $(b,r_0)$. See if the prior seed requires $M+1$ or more divisions this seed must require $M$ or more divisions. Hence it must satisfy $b\geq F_{M+2}, r_0\geq F_{M+1}$. Note the former condition is part of our statement.
BWOC suppose $a<F_{M+3}$. Then $r_0=a-bq\leq a-b<F_{M+3}-F_{M+1}=F_{M+2}$. Thus the above criterion is not met and hence we must have $a\geq F_{M+3}$.

Hence we have proven both $a\geq F_{M+3}, b\geq F_{M+2}$ and thereby proven the statement holds for $N\leq M+1$. Thus, by induction, the statement holds for all $N\in \mathbb N$.

```python
def gcd(a,b):
    """
        Returns the greatest common divisor gcd(a,b)
        using the Euclidean algorithm.
    """
    r0 = a
    r1 = b
    while r1 != 0:
        (q,r) = divmod(r0,r1)
        r0=r1
        r1=r
    return r0

def ExtendedEuclideanAlgorithm(a,b,c):
    """
        Solves for a solution of ax+by=c and returns (x,y,gcd(a,b)), or if no solutions exist returns (F,F,gcd(a,b)).
    """

    (x0,x1) = (1,0)
    (y0,y1) = (0,1)

    r0 = a
    r1 = b
    while r1 != 0:
        (q,r) = divmod(r0,r1)
        #print(r0, "=", q, "*", r1, "+", r)
        r0=r1
        r1=r

        (x2,y2) = (x0 - q * x1, y0 - q * y1)
        #print(x2, "*", a, "+",y2, "*", b, "=", r)
        (x0,y0)=(x1,y1)
        (x1,y1)=(x2,y2)
    
    # we have found x,y for ax+by=gcd(a,b)=r0. Thus if possible scale to c=q*r0.
    (q,r)=divmod(c,r0)
    if r != 0:
        return (False,False, r0)
    # valid solution found.
    (x,y) = (q*x0, q*y0)
    
    # simplify the solution a bit using the hom. solution.
    # criterion for 'best' outputs here may be useful / how to get such 'best' outputs quickly?
    
    if abs(x)>abs(y):
        (q,r)=divmod(x,b)
        return (r,y+q*a,r0)
    else:
        (q,r)=divmod(y,a)
        return (x+q*b,r,r0)

# Hence solving the given particular question.
a=993319
b=999331
c=939193
(x,y,g)=ExtendedEuclideanAlgorithm(a,b,c)
print(x, "*",a,"+",y,"*",b,"-",c,"= 0")
```
