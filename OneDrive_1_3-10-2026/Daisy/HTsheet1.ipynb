```python

def gcd(x,y): 
 #gcd of x and y using Euclid's method
    if x<y:
     x,y=y,x
    r=x-y*(x/y)
    x=y
    y=r
    if r==0:
      return(x)

# extended euclidean algorithm
def extended_gcd(a, b):
   x,y, u,v = 0,1, 1,0
   while a != 0:
        q, r = b//a, b%a
        m, n = x-u*q, y-v*q
        b,a, x,y, u,v = a,r, u,v, m,n
        gcd = b
   return gcd, x, y

def solution(a, b, c): #solves ax+by=c

    # get gcd(a, b) and coefficients
    gcd, x0, y0 = extended_gcd(a, b)

    # check if solution exists
    if c % gcd != 0:
        return("There is no solution")

    # scale the solution to match c
    f = c // gcd
    x = x0 * f
    y = y0 * f

    return x, y

print("The solution to the equation 993319x+999331y=939193 is x,y=")
print(solution(993319,999331,939193))
```

```python
def rsa_private_key(p, q, e):
   #computes the RSA private key d given primes p, q and public key e
   lambda_n = ((p - 1) * (q - 1)) // gcd(p-1,q-1)
   gcd_val, d, _ = extended_gcd(e, lambda_n)
   if gcd_val != 1:
       return("None")
   else:
     d = d % lambda_n
     return d
   
print("d=")
print(rsa_private_key(137,173,7))
```
