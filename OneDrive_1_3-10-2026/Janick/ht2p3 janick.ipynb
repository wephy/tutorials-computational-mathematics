```python
import sympy as sp

x = sp.symbols('x')
funcs = {"f(x) = sin(x)sinh(x)":sp.sin(x)*sp.sinh(x),
        "f(x) = x - sin(x)":x - sp.sin(x), 
        "f(x) = x*cos(x)":x*sp.cos(x)}



for function in funcs:
  f_diff = sp.diff(funcs[function], x)
  
  f = sp.lambdify(x, funcs[function])
  f_diff = sp.lambdify(x, f_diff)
  x_0 = (1/2)
  count = 0

  linear_factors = []
  cubic_factors = []

  while abs(f(x_0)) > 1e-8:
    x_1 = x_0 - f(x_0)/f_diff(x_0)
    count += 1

    linear_factors.append(x_1/x_0)
    cubic_factors.append((x_1/(x_0**3)))
    x_0 = x_1
    
  print(f"Function: {function}, Iterations: {count}")
  print(f"Linear factors: {[float(x) for x in linear_factors]}")
  if function == "f(x) = x*cos(x)":
    print(f"Cubic factors: {[float(x) for x in cubic_factors]}")
```

In the above, we can see that functions 1 and 2 converge linearly, as e_n+1/e_n converges to some C. Wheras for function 3  e_n+1/(e_n)^3 converges to a limit so it has cubic order, as we would predict by Q2.
