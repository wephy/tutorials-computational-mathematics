```python
import numpy as np
from sympy import symbols, Matrix

x = symbols('x')
y = symbols('y')

system = Matrix([x**2 - y - 1,
            x - y**2 + 1])

jacobian = system.jacobian(Matrix([x, y]))

print(jacobian)

def F(x):
    return np.array([
        x[0]**2 - x[1] - 1,
        x[0] - x[1]**2 + 1
    ])

def J(x):
    return np.array([
        [2*x[0], -1],
        [1, -2*x[1]]
    ])

def newton_iterator(F, J, x0, tol):
    x = np.array(x0)
    iteration = 0
    while np.linalg.norm(F(x)) >= tol:
        iteration += 1
        x = x + np.linalg.solve(J(x), -F(x))
    print(f"Converged in {iteration} iterations.")
    return x

print(newton_iterator(F, J, [1.0, 1.0], 1e-8))
print(newton_iterator(F, J, [-67, 67], 1e-8))
print(newton_iterator(F, J, [67, -67], 1e-8))
```

So the x = y = golden ratio, x=-1,y=0 and x=0 y=-1 are 3 pairs of solutions

Q5: Global minimum is when E=0 at (x,y) = (1,1)

```python
def grad(x):
    return np.array([
        -2*(1-x[0]) - 400*x[0]*(x[1] - x[0]**2),
        200*(x[1] - x[0]**2)
    ])

def hessian(x):
    return np.array([
        [2 - 400*(x[1] - x[0]**2) + 800*x[0]**2, -400*x[0]],
        [-400*x[0], 200]
    ])

print(newton_iterator(grad, hessian, [2, 2], 1e-8))
print(newton_iterator(grad, hessian, [10, 10], 1e-8))
print(newton_iterator(grad, hessian, [100, 100], 1e-8))
print(newton_iterator(grad, hessian, [102350, -25100], 1e-8))

# wow very efficient
```

```python
def I(x,alpha=1000):
    return np.array([
        [alpha, 0],
        [0,alpha]
    ])

print(newton_iterator(grad, I, [2, 2], 1e-8))

def I(x,alpha=2000):
    return np.array([
        [alpha, 0],
        [0,alpha]
    ])

print(newton_iterator(grad, I, [2, 2], 1e-8))

def I(x,alpha=500):
    return np.array([
        [alpha, 0],
        [0,alpha]
    ])

print(newton_iterator(grad, I, [2, 2], 1e-8))

# is there a better approach to keep code shorter?
```

1000 converges faster than 2000 because it has a smaller step size. The step size of alpha=500 is too large so it cant converge to the desired accuracy and becomes unstable
