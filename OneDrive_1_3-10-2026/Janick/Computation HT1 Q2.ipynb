Implementation of the extended euclidean algorithm to determine solutions if they exist, starting parameters as in the question

```python
a = 993319
b = 999331
c = 939193

def gcd(x, y):
    r0,r1 = x,y
    s0,s1 = 1,0
    t0,t1 = 0,1

    while r1:
        q = (r0 - r0 % r1)//r1
        r0,r1 = r1, r0 % r1
        s0,s1 = s1, s0 - q*s1
        t0,t1 = t1, t0 - q*t1
    return [r0,s0,t0]


result = gcd(a, b)

# The above will solve ax + by = 1, so needs to be scaled by a factor of c
print(f"{a}*{result[1]*c} + {b}*{result[2]*c} = {c}")
```
