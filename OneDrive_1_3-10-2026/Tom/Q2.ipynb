```python
def Euclid_extended(a, b, c):
	x = [1, 0]
	y = [0, 1]
	r = [a, b]

	while r[1] > 0:
		q = r[0]//r[1]
		new_r = r[0] - q*r[1]
	
		r = [r[1], new_r]

		
		new_x, new_y = x[0] - q*x[1], y[0] - q*y[1]
		x, y = [x[1], new_x], [y[1], new_y]

		#print(f"{a}*{new_x} + {b}*{new_y} = {new_r}")

	if not c:	return(r[0]) # This just outputs the hcf if c is left blank
	elif c%r[0] != 0:	print("No solutions!")
	else:	return([c*x[0]//r[0], c*y[0]//r[0]])

sol_1 = Euclid_extended(993319, 999331, 939193)
print(f"993319*{sol_1[0]} + 999331*{sol_1[1]} = 939193")
```

```python
def find_lambda(p, q):	return((p-1)*(q-1)//Euclid_extended(p-1, q-1, ""))

def d(p, q, e):
	l = find_lambda(p, q)
	r = 1
	while (e*r)%l != 1:	r+=1
	return r

print(f"d = {d(173, 137, 7)}")
```
