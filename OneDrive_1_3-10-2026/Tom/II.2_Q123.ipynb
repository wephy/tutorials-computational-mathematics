### Q1

Let a, b be the numbers to be multiplied.

Then for each digit of a, it has to be multiplied by every digit from b, and added to a running total.

So there are n digits * (n multiplications + n additions) = $2n^2$ operations

Hence the time complexity is $O(n^2)$

### Q2

a) $\displaystyle x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} = g(x_n)$

So $\displaystyle x_{n+1} = g(x_n) = g(\alpha) + (x_n - \alpha)g'(\alpha) + \frac{(x_n - \alpha)^2 g''(\zeta_n)}{2}$ for some $\zeta_n \in [x_n,\alpha]$

$\displaystyle = \alpha + (x_n - \alpha)g'(\alpha) + \frac{(x_n - \alpha)^2 g''(\zeta_n)}{2}$ for some $\zeta_n \in [a,b] \supset [x_n, \alpha]$

$\displaystyle g'(x) = \frac{f(x)f''(x)}{f'(x)^2}$, so $g'(\alpha) = 0$ because $f(\alpha) = 0$

Then $\displaystyle x_{n+1} = \alpha + \frac{(x_n - \alpha)^2 g''(\zeta_n)}{2}$ for some $\zeta_n \in [a,b]$

So $\displaystyle \frac{|x_{n+1} - \alpha|}{|x_n - \alpha|^2} ≤ \frac{|\frac{1}{2}(x_n - \alpha)^2 g''(\zeta^*)|}{|x_n - \alpha|^2}$ where $|g''|$ reaches a maximum at $\zeta^* \in [a,b]$

$\displaystyle ≤ \frac{1}{2} |g''(\zeta^*)|$

Hence the method converges at least quadratically <br> <br><br><br><br>

b) $\displaystyle x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$

So $\displaystyle \frac{|x_{n+1} - \alpha|}{|x_n - \alpha|} = \frac{|f'(x_n)(x_n - \alpha) - f(x_n)|}{|f'(x_n)(x_n - \alpha)|} = |1 - \frac{f(x_n)}{f'(x_n)(x_n - \alpha)}|$

By l'Hôpital's rule, $\displaystyle \lim\limits_{x_n \to \alpha} \frac{f(x_n)}{f'(x_n)(x_n - \alpha)} = \lim\limits_{x_n \to \alpha} \frac{f'(x_n)}{f''(x_n)(x_n - \alpha) + f'(x_n)} = \lim\limits_{x_n \to \alpha} \frac{f''(x_n)}{f'''(x_n)(x_n - \alpha) + 2f''(x_n)}$

$\displaystyle = \lim\limits_{x_n \to \alpha} \frac{1}{\frac{f'''(x_n)(x_n - \alpha)}{f''(x_n)} + 2} = \frac{1}{2}$

Hence the method converges linearly, with rate $\frac{1}{2}$ <br><br><br><br><br>

c) As above, $\displaystyle \lim\limits_{n \to \infty} \frac{|x_{n+1} - \alpha|}{|x_n - \alpha|} = \lim\limits_{x_n \to \alpha} \frac{f''(x_n)}{f'''(x_n)(x_n - \alpha) + 2f''(x_n)}$

<br>

$\displaystyle = \lim\limits_{x_n \to \alpha} \frac{f'''(x_n)}{f''''(x_n)(x_n - \alpha) + 3f'''(x_n)}$ by l'Hôpital

<br>

$\displaystyle = \lim\limits_{x_n \to \alpha} \frac{1}{\frac{f''''(x_n)(x_n - \alpha)}{f'''(x_n)} + 3} = \frac{1}{3}$

So the rate of convergence is $\frac{1}{3}$ <br><br><br><br><br>

d) $\displaystyle x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)} = g(x_n)$

So $\displaystyle x_{n+1} = g(x_n) = g(\alpha) + (x_n - \alpha)g'(\alpha) + \frac{(x_n - \alpha)^2 g''(\alpha)}{2} + \frac{(x_n - \alpha)^3 g'''(\zeta_n)}{6}$ for some $\zeta_n \in [x_n,\alpha]$

$\displaystyle = \alpha + (x_n - \alpha)g'(\alpha) + \frac{(x_n - \alpha)^2 g''(\alpha)}{2} + \frac{(x_n - \alpha)^3 g'''(\zeta_n)}{6}$ for some $\zeta_n \in [a,b] \supset [x_n, \alpha]$

<br>

$\displaystyle g'(x) = \frac{f(x)f''(x)}{f'(x)^2}$, so $g'(\alpha) = 0$ because $f(\alpha) = 0$

$\displaystyle g''(x) = \frac{[f'(x)f''(x)+f(x)f'''(x)][f'(x)^2]-[f(x)f''(x)][2f'(x)f''(x)]}{f'(x)^4}$, so $g''(\alpha) = 0$ because $f(\alpha) = 0 = f''(\alpha)$

<br>

Then $\displaystyle x_{n+1} = \alpha + \frac{(x_n - \alpha)^3 g'''(\zeta_n)}{6}$ for some $\zeta_n \in [a,b]$

So $\displaystyle \frac{|x_{n+1} - \alpha|}{|x_n - \alpha|^3} ≤ \frac{|\frac{1}{6}(x_n - \alpha)^3 g'''(\zeta^*)|}{|x_n - \alpha|^3}$ where $|g'''|$ reaches a maximum at $\zeta^* \in [a,b]$

$\displaystyle ≤ \frac{1}{6} |g'''(\zeta^*)|$

Hence the method converges at least cubically

### Q3

```python
import matplotlib.pyplot as plt, numpy as np, mpmath as mp
mp.mp.dps = 100

def N_R(f, dfdx, x0):
    x = x0
    while True:
        x = x - f(x) / dfdx(x)
        yield(x)

a = N_R(lambda x: mp.sin(x)*mp.sinh(x), lambda x: mp.cos(x)*mp.sinh(x) + mp.sin(x)*mp.cosh(x), mp.mpf("0.5"))
b = N_R(lambda x: x - mp.sin(x), lambda x: 1 - mp.cos(x), mp.mpf("0.5"))
c = N_R(lambda x: x*mp.cos(x), lambda x: mp.cos(x) - x*mp.sin(x), mp.mpf("0.5"))

a_values, b_values, c_values = [0.5], [0.5], [0.5]

fig, ax = plt.subplots()
ax.set_xlabel("n")
ax.set_ylabel(r"$x_n$")
plt.gcf().set_dpi(300)
plt.tight_layout()

fig2, ax2 = plt.subplots()
ax2.set_yscale("log")
ax2.set_xlabel("n")
ax2.set_ylabel(r"$|x_n|$ (log scale)")
plt.gcf().set_dpi(300)
plt.tight_layout()

for i in range(3):
    sequence = [a,b,c][i]
    values = [a_values,b_values,c_values][i]
    while abs(values[-1]) > 10**-8: values.append(next(sequence))

    # Now plot that sequence on the graph
    values = np.array(values)
    ax.plot(range(len(values)), values, label=[r"$\sin{(x)}\sinh{(x)}$", r"$x-\sin{(x)}$", r"$x\cos{(x)}$"][i])
    ax2.plot(range(len(values)), abs(values), label=[r"$\sin{(x)}\sinh{(x)}$", r"$x-\sin{(x)}$", r"$x\cos{(x)}$"][i])

ax.legend()
ax2.legend()
plt.show()
```

a) $f'(x) = \cos{x}\sinh{x} + \sin{x}\cosh{x} \implies f'(\alpha) = 0$

$f''(x) = \sin{x}(-\sinh{x} + \sinh{x}) + \cos{x}(\cosh{x} + \cosh{x}) = 2\cos{x}\cosh{x} \implies f''(\alpha) = 2$

so the convergence is linear, with rate of convergence $\frac{1}{2}$

b) $f'(x) = 1 - \cos{x} \implies f'(\alpha) = 0$

$f''(x) = \sin{x} \implies f''(\alpha) = 0$

$f'''(x) = \cos{x} \implies f'''(\alpha) = 3$

so the convergence is linear, with rate of convergence $\frac{1}{3}$

c) $f'(x) = \cos{x} - x\sin{x} \implies f'(\alpha) = 1$

$f''(x) = -x\cos{x} - 2\sin{x} \implies f''(\alpha) = 0$

so the convergence is at least cubic
