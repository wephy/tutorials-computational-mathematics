```python
import numpy as np
```

```python
def newton(F, J, x0, tol, n=0):
  
  if np.linalg.norm(F(x0)) < tol:
    return (x0, n)
  
  dx = np.linalg.solve(J(x0), -F(x0))
  return newton(F, J, x0 + dx, tol, n+1)
```

```python
F = lambda x: np.array([x[0]**2 - x[1] - 1, x[0] - x[1]**2 + 1])
# Then solutions of the system are roots of this F

# Jacobian
J = lambda x: np.array([
  [2*x[0], -1],
  [1, -2*x[1]]
])

print("By inspection: (x, y) = (1, 0) or (0, -1)")
print("By Newton iteration:")
print(newton(F, J, np.array([0.0, 0.0]), 1e-8)[0])
print(newton(F, J, np.array([1.0, 1.0]), 1e-8)[0])
```

**5.** (a) The global minimum is at $E(1,1)=0$.

(b) $\nabla E = (2x-2+400x^3-400xy, 200y-200x^2)$.

```python
startings = [
  np.array([2, 2]),
  np.array([10, 10]),
  np.array([100, 100]),
  np.array([1e6, 1e6])
]

F = lambda x: np.array([
  2*x[0] - 2 + 400*x[0]**3 - 400*x[0]*x[1],
  200*(x[1]-x[0]**2)
])

J = lambda x: np.array([
  [2 + 1200*x[0]**2 - 400*x[1], -400*x[0]],
  [-400*x[0], 200]
])

for start in startings:
  root = newton(F, J, start, 1e-8)
  print(f"Started at {start} and found root {root[0]} with {root[1]} iterations")
```
