**1.** We claim that if $\gcd(a,b)$ for some $a>b$ requires $N$ divisions to compute, then $a \geqslant F_{N+2}$ and $b \geqslant F_{N+1}$.

We note first that the statement does not hold for $N=0$, as the only way for no divisions to be necessary is when $b=0$, but then $b \ngeqslant 1 = F_1$. So our base case will be $N=1$. Then we know $b>0$, so $b \geqslant 1 = F_{2}$.

For the inductive hypothesis, assume that for all $a>b$ where $\gcd(a,b)$ requires $N$ divisions, we have $a \geqslant F_{N+2}$ and $b \geqslant F_{N+1}$.
Now let $(a,b)$ such that $\gcd(a,b)$ requires $N+1$ divisions. After the first division, we have $(b, a-qb)$ where $q>0$ and $0 \leqslant a-qb < b$. Finding $\gcd(b, a-qb)$ requires $N$ further divisions, so we know by the inductive hypothesis that $b \geqslant F_{N+2}$ and $a-qb \geqslant F_{N+1}$. But then $qb \geqslant b \geqslant F_{N+2}$, so $a = (a-qb) + qb \geqslant F_{N+1} + F_{N+2} = F_{N+3}$. Thus when $\gcd(a,b)$ requires $N+1$ divisions we must have $a \geqslant F_{N+3}$ and $b \geqslant F_{N+2}$. $\qquad\square$

**2.** (a)

```python
def ext_euler(a, b, c):
  if b > a: # We need a >= b
    return ext_euler(b, a, c)[::-1]
  r0, r1, s0, s1, t0, t1 = a, b, 1, 0, 0, 1
  while r1 > 0:
    q = r0 // r1
    r0, r1 = r1, r0 - q * r1
    s0, s1 = s1, s0 - q * s1
    t0, t1 = t1, t0 - q * t1
  assert not c % r0 # Check that the gcd divides the target value
  return s0*c//r0, t0*c//r0

a = 993319
b = 999331
c = 939193
x, y = ext_euler(a, b, c)
assert a * x + b * y == c # Check the solution
print("x =", x)
print("y =", y)
print(f"{a}*{x} + {b}*{y} = {c}")
```

```python
def gcd(a, b): # Euler algorithm (could also be done recursively)
  a, b = max(a, b), min(a, b) # Ensure a > b
  while b > 0:
    a, b = b, a % b
  return a

p = 137
q = 173
e = 7

L = (p-1) * (q-1) // gcd(p-1, q-1) # Carmichael totient func
assert 1 < e < L
assert gcd(e, L) == 1

# Solving de = 1 mod L is equivalent to solving de + xL = 1

d = ext_euler(e, L, 1)[0]
print("Private key:", d)
```

```python
m = 21051
n = p * q
c = (m**e)%n
print("Ciphertext:", c)
m1 = (c**d)%n
print("Plaintext:", m1)
assert m1 == m
```
