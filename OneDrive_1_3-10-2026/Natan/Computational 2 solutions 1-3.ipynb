**2.** (a) Recall that Newton's method is equivalent to fixed point iteration of the function $g(x) = x - \frac{f(x)}{f'(x)}$. So consider the sequence $x_{i+1}=g(x_i)$. The Taylor series for $g$ around some point $\xi$ is $g(x) = g(\xi) + (x-\xi)g'(\xi) + \frac{(x-\xi)^2}2g''(\zeta)$ for some $\zeta$ between $x$ and $\xi$.


Let $x_n \to \alpha$ as $n\to\infty$. Then consider the Taylor series evaluated around $\alpha$. We have $g(x) = g(\alpha) + (x-\alpha)g'(x) + \frac{(x-\alpha)^2}2g''(\zeta)$. Now substitute $x$ for $x_i$, then note that $g(x_i) = x_{i+1}$, and $g(\alpha) = \alpha$. Then we have
$$\begin{align*}
  x_{i+1} &= \alpha + (x_i-\alpha)g'(\alpha) + \frac{(x_i-\alpha)^2}2g''(\zeta) \\
  |x_{i+1} - \alpha| &= |(x_i-\alpha)g'(\alpha) + \frac{(x_i-\alpha)^2}2g''(\zeta)| \\
  &\leqslant |(x_i-\alpha)|\ |g'(\alpha)| + \frac12|x_i-\alpha|^2\ |\sup_{s\in(x_i, \alpha)}g''(s)|
\end{align*}$$

Let $f'(\alpha)\neq 0$. Then $g'(\alpha) = 1 - \frac{f'(\alpha)^2 - f(\alpha)f''(\alpha)}{f'(\alpha)^2} = \frac{f(\alpha)f''(\alpha)}{f'(\alpha)^2} = 0$.
Then $\frac{|x_{i+1} - \alpha|}{|x_i-\alpha|^2} = \frac12 |\sup_{s\in(x_i, \alpha)}g''(s)|$. Thus we have quadratic convergence.

```python
import sympy as sp
import numpy as np

def derivatives(_f, x):
  _fp = sp.diff(_f, x)
  _fpp = sp.diff(_fp, x)
  f = sp.lambdify(x, _f, 'numpy')
  fp = sp.lambdify(x, _fp, 'numpy')
  fpp = sp.lambdify(x, _fpp, 'numpy')
  return f, fp, fpp
```

```python
def newton(f, fp, x0):
  x = x0
  while True:
    yield x
    x -= f(x) / fp(x)
```

(a)

```python
x = sp.symbols("x")
f, f1, f2 = derivatives(sp.sin(x) * sp.sinh(x), x)

rootgen = newton(f, f1, 0.5)
roots = [next(rootgen)]
while abs(f(roots[-1])) >= 1e-8:
  roots.append(next(rootgen))
for i in range(100): # Get a much more accurate root for error calculations
  a = next(rootgen)
print("Final root α: ", round(a, 8))
print("f(α) = ", f(a))
print("f'(α) = ", f1(a))
print("f''(α) = ", f2(a))
print("Successive errors:")
print("\n".join(str(r - a) for r in roots))
print("Linear ratios:")
for i in range(len(roots) - 1):
  print(abs(roots[i+1] - a) / abs(roots[i] - a))
```

Here $f'(\alpha)=0$ and $f''(\alpha)\neq0$, and the convergence is linear as expected.

Here $f'(\alpha)=f''(\alpha)=0$, and the convergence is linear with rate $\frac23$ as expected.
