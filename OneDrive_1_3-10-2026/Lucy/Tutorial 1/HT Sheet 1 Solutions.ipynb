### Question 2a.

For this question I am implementing the extended Euclidean algorithm to solve the equation ax+by=c
without using excessive memory.
The program keeps track of the most recent two remainders, and the last two entries in the series x and y.

```
import math
a=993319
b=999331
c=939193
```

```
def solve_equation(a, b, c):
    g=math.gcd(a, b)
    if c%g==0:
        x=[1, 0]
        y=[0, 1]
        r=[a, b]
        r_temp=a%b
        while r_temp!=0:
            q=r[0]//r[1]
            x_temp=x[0]-q*x[1]
            y_temp=y[0]-q*y[1]
            r[0],r[1]=r[1],r_temp
            x[0],x[1]=x[1],x_temp
            y[0],y[1]=y[1],y_temp
            r_temp=r[0]%r[1]
        factor=c//g
        return[x[1]*factor,y[1]*factor]
    return("no solution")
```

```
solution=(solve_equation(a, b, c))
print("x=",solution[0],"y=",solution[1])
```
