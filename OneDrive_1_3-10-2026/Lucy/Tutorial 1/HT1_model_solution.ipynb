## Question 1.

1. Prove by induction that if Euclid’s algorithm requires $N$ divisions to compute the greatest
common divisor of the pair $a > b, a, b \in N$, then $a \ge F_{N+2}$ and $b \ge F_{N+1}$, where $F_i$ denotes
the $i^\text{th}$ Fibonacci number ($F_1 = 1, F_2 = 1, F_{i+2} = F_{i+1} + F_i$)

1. Base case, $P(1)$:

If $\gcd(a, b)$ requires 1 division for $a > b$, then $\gcd(a, b) = b$. We have $b \ge F_{1+1} = 1$, and $a \ge 2 = F_{1+2}$

2. Assume the induction hypothesis for a particular $k$, $P(k)$:

If $\gcd(a, b)$ requires $k$ divisions for $a > b$, then $a \ge F_{k+2}$ and $b \ge F_{k+1}$.

3. Show that for every $\ge 1$, $P(k) \implies P(k+1)$

If $\gcd(c, d)$ requires $k+1$ divisions for $c > d$, we have $\gcd(d, (c \mod d)) = k$, and so $d \ge F_{(k)+2}$ and $(c \mod d) \ge F_{(k)+1}$.

We immediately have $d \ge F_{(k+1) + 1}$.

Recall that $F_{(k+1) + 2} = F_{(k+1) + 1} + F_{(k+1)}$, and since $c \ge (c \mod d) + d$, we have $c \ge  F_{(k)+1} + F_{(k+1) + 1} = F_{(k+1)+ 2}$.

3. Conclusion:

Since both the base case and the induction step have been proved as true, by mathematical induction the statement $P(n)$ holds for every natural number $n \ge 1.$

### (a)

The Extended Euclidean Algorithm finds $\gcd(a,b)$ and solves $ax+by = \gcd(a,b)$

Let's walk through an example:

$a = 30 = 1a + 0b$

$b = 25 = 0a + 1b$

$30 \mod 25 = 5 = 1a - 1b \quad \leftarrow b$ quantity can be achieved by divmod

$25 \mod 5 = 0 = 1b - 5 * (1a - 1b) = -5a + 6b$

$-5a + 6b = 0$, but we want the penultimate step, $1a - 1b = \gcd(a, b)$

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

```python
import numpy as np
```

```python
a = 30
b = 25
x = np.array([1, 0])
y = np.array([0, 1])
q, r = divmod(a, b)
x, y = y, x - q * y
print(x, y)
a, b = b, r
q, r = divmod(a, b)
x, y = y, x - q * y
print(x, y)
print(r, a, b)
```

```python
def extended_gcd(a, b):
    # x and y represent how much of the original [a, b] are in the current two arguments
    x = np.array([1, 0], dtype=object)
    y = np.array([0, 1], dtype=object)
    while True:
        q, r = divmod(a, b)
        # left argument inherents right argument, and right argument updates accordingly to the subtraction
        x, y = y, x - q * y
        a, b = b, r
        if r == 0:
            return x
```

```python
gcd(993319, 999331)
```

If their gcd is 1, it means we can easily solve for $993319x + 999331y = 1$

```python
def solver(a, b, c):
    g = gcd(a, b)
    q, r = divmod(c, g)
    if r == 0:
        return q * extended_gcd(a, b)
    else:
        return False
```

```python
solver(993319, 999331, 939193)
```

Our answer: $993319 \cdot 347668586354 + 999331 \cdot -345577003543 = 939193$

x = 347668586354

y = -345577003543

```python
def carmichael(p, q):
    return (p - 1) * (q - 1) // gcd(p - 1, q - 1)
```

### Extension

If I compute $s \equiv m^d \mod n$ and give you $s$, can you verify that I have a private key?

#### Solution

$s \equiv m^d \mod n$

$(m^d)^e = m^{(de)} = (m^e)^d$

$(m^e)^d \equiv c^d \mod n$

$c^d \equiv m \mod n$

So we have:

$s^e \equiv m \mod n$

And as $e$ is a public key, you can verify ownership of a private key with $s$.
