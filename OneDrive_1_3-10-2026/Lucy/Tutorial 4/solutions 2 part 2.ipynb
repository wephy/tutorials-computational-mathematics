```python
import numpy as np
#make use of numpy arrays
```

```python
def newton(F, J, x0, tol):
    x1=x0 - np.linalg.solve(J(x0), F(x0))
    while np.linalg.norm(F(x1))>tol:
        x0=x1
        x1=x0 - np.linalg.solve(J(x0), F(x0)) 
    
    return x1
```

```python
def F(xs):
    x=xs[0]
    y=xs[1]
    return np.array([x**2-y-1,x-y**2+1])
```

```python
def J(xs):
    x=xs[0]
    y=xs[1]
    return np.array([[2*x, -1],
                    [1, -2*y]])
```

```python
tol=10**-8
solutions=[]
#want to start from a range of x0's
for m in range (0, 100):
    for n in range (0, 100):
        for i in range(0,2):
            for j in range(0, 1):
                try:
                    x0=np.array([(m*(-1)**i)/2,(n*(-1)**j)/2])
                    s=newton(F, J, x0, tol)
                    s=np.round(s,7)
                    if not any(np.allclose(s, sol) for sol in solutions):
                        solutions.append(s)
                except np.linalg.LinAlgError:
                    pass
```

```python
print(solutions)
```
