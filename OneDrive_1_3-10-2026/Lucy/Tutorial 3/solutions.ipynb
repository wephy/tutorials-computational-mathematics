### Q3a


```
import mpmath as mp
mp.mp.dps=50
```

```
def f(x):
    return mp.sinh(x)*mp.sin(x)
```

```
def fd(x):
    return mp.sinh(x)*mp.cos(x)+mp.cosh(x)*mp.sin(x)
```

```
x=mp.mpf(0.5)
```

```
while abs(f(x))>(mp.mpf(10**(-8))):
    x=x-(f(x))/(fd(x))
 #   print(x)
```

```
print(x)
```
