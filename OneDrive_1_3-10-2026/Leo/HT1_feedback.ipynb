```python
def Extended_Euclidean_algorithm(a,b,c=1,output=0):
    #initialise the first 2 terms of each list according to the algorithm
    r=[a,b]
    x=[1,0]
    y=[0,1]
    q=[]

    #r[1] is remainder after each division so when =0 we have reached the end of the algorithm

    while r[1]!=0:
        #finds the modulo and quotient at once.
        q=divmod(r[0],r[1])
        #writes the new modulo into our remainder list and removes the one that now doesn't need to be stored
        r.append(q[1])
        r.remove(r[0])
        #writes the next set of x,y so that ax+by=r while removing now unused terms
        x.append(x[0]-q[0]*x[1])
        x.remove(x[0])
        y.append(y[0]-q[0]*y[1])
        y.remove(y[0])
        #r[0] is gcd(a,b) so r[0] needs to be a factor of c for a solution to exist 
        if c%r[0]==0:
            #scales by c to get x and y then adds 0 to find a "sensible decision"
            x=int(x[0]*c/r[0])
            simplify=divmod(x,b)[0]
            x-=b*simplify
            y=int(y[0]*c/r[0])+a*simplify
            # allows us to read out our variables for 2b
            _output=(f'{a}*{x}+{b}*{y}={c} x={x},y={y}',x,y,r[0])
        else:
            _output=(f'No Integer Solutions',x,y,r[0])
    return(_output[output])
#allows user to input our desired values of a,b,c
print("Enter the diophantine equation. (ax+by=c to be inputed as a,b,c)")
_input=eval("["+input()+"]")
print(Extended_Euclidean_algorithm(_input[0],_input[1],_input[2]))
#Algorithm runs in 0.0s inputting digits is what leads to long times
```

```python
def Extended_Euclidean_algorithm_fix(a, b, c=1, output=0):
    r = [a, b]
    x = [1, 0]
    y = [0, 1]

    while r[1] != 0:
        q = divmod(r[0], r[1])
        
        # We must update lists before checking the condition
        r.append(q[1])
        r.remove(r[0])
        
        x.append(x[0] - q[0] * x[1])
        x.remove(x[0])
        
        y.append(y[0] - q[0] * y[1])
        y.remove(y[0])

    # Moved outside the loop: r[0] is now the final GCD
    if c % r[0] == 0:
        factor = c // r[0]
        # Calculate the specific solution for c
        x_val = x[0] * factor
        y_val = y[0] * factor
        
        # Your "sensible decision" reduction logic
        simplify = (x_val // (b // r[0]))
        x_final = x_val - (b // r[0]) * simplify
        y_final = y_val + (a // r[0]) * simplify
        
        _output = (f'{a}*{x_final}+{b}*{y_final}={c} x={x_final},y={y_final}', x_final, y_final, r[0])
    else:
        _output = ('No Integer Solutions', None, None, r[0])
        
    return _output[output]

print(Extended_Euclidean_algorithm_fix(993319, 999331, 939193))
```
