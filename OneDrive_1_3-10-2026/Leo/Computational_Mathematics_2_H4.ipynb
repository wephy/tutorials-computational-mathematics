We will assume single digit multiplication/ addition have $O(1)$. We have $n^2$ multiplications as we times each digit in the first number with each digit of the 2nd number. For addition take the worst case (i.e. every multiplication requires a "carry" and then every sum needs a carry). we are then adding 2n n-digit numbers (ignoring 0s on the right hand side of the number). so We have 2n*n additions (ignoring carries). we then need to add the carries which can be on every digit bar the units place so another 2n-1 additions (note the leftmost addition doesn't have a carry so remove that one from worst case). This leads us to time complexity $O(n^2)$ as $n^2+2n^2+2n-1 \leq (2n+1)^2$

$Q2a)$
Let $M(x^n)$ be defined as the sum of polynomial terms of at least degree n.

Let Newton's algorithm for f(x) be described as $g(x)=x-\frac{f(x)}{f'(x)}, x_{i+1}=g(x_i)$.

$ g'(x)=1-\frac{f'(x)^2-f(x)f''(x)}{f'(x)^2}=\frac{f(x)f''(x)}{f'(x)^2}$ so $g'(\alpha)=0$

$\frac{|x_{i+1}-\alpha|}{|x_i-\alpha|^2}=\frac{|-\alpha+g(\alpha)+g'(\alpha)(x_i-\alpha)+M((x_i-\alpha)^2)|}{|x_i-\alpha|^2}=\frac{|M((x_i-\alpha)^2)|}{|x_i-\alpha|^2} \leq \frac{M(|x_i-\alpha|^2)}{|x_i-\alpha|^2}=M(1)$ by triangle inequality.

So if $\forall i \space x_i \neq \alpha $. then $\lim_{i\to\infty} \frac{|x_{i+1}-\alpha|}{|x_i-\alpha|^2} \to 1$ or $\infty$ so order of covergence at least 2.

$Q2b)$
$\frac{|x_{i+1}-\alpha|}{|x_i-\alpha|}=\frac{|-\alpha+g(\alpha)+g'(\alpha)(x_i-\alpha)+M((x_i-\alpha)^2)|}{|x_i-\alpha|}=\frac{|g'(\alpha)(x_i-\alpha)+M((x_i-\alpha)^2)|}{|x_i-\alpha|}$

For $\lim_{x\to\alpha}g'(x)$ We can apply l'Hopital's rule to this twice as we have met all the conditions. so $\lim_{x\to\alpha}g'(x)= \lim_{x\to\alpha} \frac{f(x)f'''(x)+f'(x)f''(x)}{2f'(x)f''(x)}= \lim_{x\to\alpha} \frac{2f'(x)f'''(x)+f(x)f''''(x)+(f''(x))^2}{2(f''(x))^2+2f'(x)f'''(x)}=\frac{(f''(\alpha))^2}{2(f''(\alpha))^2}=\frac{1}{2}$

So $\frac{x_{i+1}-\alpha}{x_i-\alpha}=\lim_{i\to\infty}\frac{0.5(x_i-\alpha)+M((x_i-\alpha)^2)}{x_i-\alpha}=\lim_{i\to\infty}(0.5+M((x_i-\alpha)))=\frac {1}{2}$ so Linear order of convergence. (as our limit is positive the absolute values are irrelevant)

$Q2c)$ Let's look at $Q2b)$ and continue with L'Hopital's Rule 2 times more:

$\lim_{x\to\alpha}g'(x)=\lim_{x\to\alpha}\frac{4f''(x)f'''(x)+3f'(x)f''''(x)+f(x)f'''''(x)}{6f''(x)f'''(x)+2f'(x)f''''(x)}=\lim_{x\to\alpha}\frac{4(f'''(x))^2+7f''(x)f''''(x)+4f'(x)f'''''(x)+f(x)f''''''(x)}{6(f'''(x))^2+8f''(x)f''''(x)+2f'(x)f'''''(x)}=\frac{2}{3}$

So $\frac{x_{i+1}-\alpha}{x_i-\alpha}=\lim_{i\to\infty}\frac{\frac{2}{3}(x_i-\alpha)+M((x_i-\alpha)^2)}{x_i-\alpha}=\lim_{i\to\infty}(\frac{2}{3}+M((x_i-\alpha)))=\frac {2}{3}$ so Linear order of convergence. (as our limit is positive the absolute values are irrelevant)


$Q2d)$ Luckily L'Hopital's Rule is not needed but quotient rule is about to become really long.
$g'(\alpha)=0$
$g''(x)=\frac{(f'(x))^2f''(x)+f(x)f'(x)f'''(x)-2f(x)f''(x)^2}{f'(x)^3}$ so $g''(\alpha)=0$
$g'''(x)=\frac{- \frac{2 \left(f'''{\left(x \right)} - \frac{3 \left(f''{\left(x \right)}\right)^{2}}{f'{\left(x \right)}}\right) f{\left(x \right)} f''{\left(x \right)}}{\left(f'{\left(x \right)}\right)^{2}} + \frac{f{\left(x \right)} f''''{\left(x \right)}}{f'{\left(x \right)}} - \frac{4 f{\left(x \right)} f''{\left(x \right)} f'''{\left(x \right)}}{\left(f'{\left(x \right)}\right)^{2}} + 2 f'''{\left(x \right)} - \frac{3 \left(f''{\left(x \right)}\right)^{2}}{f'{\left(x \right)}}} {f'{\left(x \right)}}$ so $g'''(\alpha)=\frac{2f'''(\alpha)}{f'(\alpha)}$

$\frac{x_{i+1}-\alpha}{(x_i-\alpha)^3}=\frac{-\alpha+g(\alpha)+g'(\alpha)(x_i-\alpha)+g''(\alpha)(x_i-\alpha)^2+g'''(\alpha)(x_i-\alpha)^3+M((x_i-\alpha)^4)}{(x_i-\alpha)^3}=\frac{2f'''(\alpha)}{f'(\alpha)}+M((x_i-\alpha)) $
so $\lim_{i\to\infty}\frac{x_{i+1}-\alpha}{(x_i-\alpha)^3}=\frac{2f'''(\alpha)}{f'(\alpha)}$ so we have cubic order convergence (as our limit is positive the absolute values are irrelevant)

```python
import sympy as sp
import numpy as np
import mpmath as mp
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(["science"])
mp.mp.dps=100 #otherwise the derivative floating pointnerrors to 0
x=sp.Symbol('x')
def Newton_Raphson(f,x0=mp.mpf(0.5), tolerance=10**-8):
    '''creates a generator that iterates the Newton-Raphson Method'''
    xn=x.subs({x:x0})
    while sp.Abs(xn.evalf())>tolerance: #we iterate continually subbing in until inside our tolerance
        g=x-f/sp.diff(f)
        xn=g.subs({x:xn})
        yield (xn) 
def plotter(xs,ys,plot_number,f):
    '''plots the graphs to find order. Heavy influence from your code last week'''
    ax=axs[plot_number]
    ax.loglog(xs,ys,'C0o-', linewidth=2, label='Data')
    grid= np.logspace(np.log10(np.float64(min(ys)))-1, 0, 100) # logarithmic list of points to draw straight line
    ax.loglog(grid, grid, 'C1--',  linewidth=2, alpha=0.5, label='Linear convergence')
    ax.loglog(grid[int(np.floor((2*len(grid))/3)):], (grid**3)[int(np.floor((2*len(grid))/3)):], 'C3--',  linewidth=2, alpha=0.5, label='Cubic convergence')
    ax.axis("equal") 
    eval(f'ax.set_title(r"${sp.latex(f)}$")')# allows us to latex our sympy function
    ax.legend(loc='upper left', facecolor='white', framealpha=1.0, frameon=True)
def _3(part,f,convergence_message,plot_number):
    '''allows nice formatting'''
    _3_iter=Newton_Raphson(f)
    _3_part_n=np.array(list(_3_iter))
    print(f"For the equation in 3{part}, α≈{list(_3_part_n)[-1]:.8e}.")
    print(f"Order of 3{part} is {convergence_message[7:]}. The expected order from {convergence_message}.")
    plotter(np.abs(_3_part_n[:-1]),np.abs(_3_part_n[1:]),plot_number,f) #the slice is there to remove first and last value, np.abs removes negatives so can log and sign is irrelevant to the problem

fig, axs = plt.subplots(1, 3, figsize=(12, 4))
_3('a',sp.sin(x)*sp.sinh(x),'2b) is Linear convergence',0)
_3('b',x-sp.sin(x),'2c) is Linear convergence',1)
_3('c',x*sp.cos(x),'2d) is Cubic convergence',2)
plt.tight_layout()
```

```python
x=sp.Symbol('x')
y=sp.Symbol('y')
def subs(f,values):
    '''creates a function to sub in so we can vectorise'''
    return np.float64(f.subs(values))
vec_subs=np.vectorize(subs)
mtrx_subs=np.vectorize(vec_subs)
def Newton(F,J,x0,tol=10**-8):
    '''applies Newton's method in the form requested'''
    F=np.array(F)
    xi=np.array(x0)
    Fxi=[2*tol]
    while np.linalg.norm(Fxi)>=tol: #finds the value of -F(x_i) and the Jacobian is at x_i
        yield (xi)
        Fxi=vec_subs(-F,{x:(xi[0]),y:(xi[1])})
        Jxi=mtrx_subs(J,({x:xi[0],y:xi[1]}))
        δx=np.linalg.solve(Jxi,Fxi)
        xi+=δx
def Jacobian(F,var=[x,y]):
    '''makes the Jacobian part of the Newton function look a bit nicer'''
    return np.array(sp.Matrix([F]).jacobian(sp.Matrix(var)))
_4b=Newton([x**2-y-1,x-y**2+1],Jacobian([x**2-y-1,x-y**2+1]),[1.,2.])
print(f'Starting with initial guess (1,2) we get [x y]={list(_4b)[-1]}.')
_4b=Newton([x**2-y-1,x-y**2+1],Jacobian([x**2-y-1,x-y**2+1]),[-1.,-2.])
print(f'Starting with initial guess (-1,-2) we get [x y]={list(_4b)[-1]}.')
```
