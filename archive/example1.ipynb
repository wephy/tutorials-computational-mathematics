```python

#
# ***
# ### Question 2 (helping Hippasus).
#

import fractions
import math
root2 = fractions.Fraction(math.sqrt(2))
history = set()
number_fractions = 0
denominator = 1

while True:
    approx = root2.limit_denominator(denominator)
    if approx not in history:
        history.add(approx)
        number_fractions = number_fractions + 1
        print(f"Approximation {number_fractions:2d}: {approx} == {float(approx)}. Error: {float(approx) - math.sqrt(2)}")

        if number_fractions == 10:
            break

    denominator = denominator + 1
```

```python

#
# ***
# ### Question 3 (helping Heron).
#

import mpmath
mpmath.mp.dps = 400
x = mpmath.mpf('1.4')   # our best guess
x_ = mpmath.mpf(1.0)  # previous guess
iteration = 0

digits = -int(mpmath.ceil(mpmath.log10(abs(x - x_))))
print(f"Iteration {iteration:2d}: x = {x} digits: {digits}", end="\n\n")

while abs(x - x_) > 1e-200:
    iteration = iteration + 1
    x_ = x
    x = 0.5 * (x + 2/x)
    digits = -int(mpmath.ceil(mpmath.log10(abs(x - x_))))
    print(f"Iteration {iteration:2d}: x = {x} digits: {digits}", end="\n\n")

#
# The iteration gets 50 digits after 6 steps, 100 digits after 7 steps, 200 digits after 8 steps.
#

#
# ***
#
```
