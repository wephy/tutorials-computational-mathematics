```python

#
# ***
# ### Question 1 (helping Eratosthenes).
#

from numpy import array

def primes(N):
    prime = array([True for n in range(N+1)])  # also include flags for 0, 1
    prime[0] = False  # 0 is not prime
    prime[1] = False  # 1 is not prime

    p = 2
    while p**2 <= N:
        if prime[p]:
            prime[2*p::p] = False

        p = p + 1

    return [n for n in range(N+1) if prime[n]]

N = 1_000_000
print(f"# prime numbers up to {N}: {len(primes(N))}")
```

```python

#
# ***
# ### Question 2 (helping de Polignac).
#

p = primes(2002)
twins = [(p, q) for (p, q) in zip(p, p[1:]) if p + 2 == q]
print("Twin primes: ")
print(twins)

```

```python

#
# ***
# ### Question 3 (helping Bode).
#

import matplotlib.pyplot as plt

# Distances from the Sun, measured in astronomical units
true_data = {"Mercury": 0.3086,
             "Venus": 0.7244,
             "Earth": 1.003,
             "Mars": 1.601,
             "Ceres": 2.87,
             "Jupiter": 4.969,
             "Saturn": 9.765,
             "Uranus": 19.63,
             "Neptune": 29.91,
             "Pluto": 40.7}

# Extract true data, and calculate Titius--Bode law
xs = list(range(10))
ys_true = true_data.values()
ys_tb = [0.4] + [0.4 + 0.3 * 2**n for n in range(9)]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(xs, ys_true, '-k', label='True distances')
ax.plot(xs, ys_tb, '--r', label='Titius--Bode law')
ax.set_title("Evaluation of the Titius--Bode law")
ax.set_ylabel("Distance from the Sun (AU)")
ax.set_xlabel("Planet")
ax.legend()
ax.set_xticks(ticks=xs, labels=true_data.keys())
plt.gcf().set_dpi(100)
plt.show()

#
# Observations: the Titius--Bode law is remarkably accurate from Mercury until Uranus,
# but is inaccurate beyond this.
#
```

```python

#
# ***
# ### Question 4 (helping Weierstrass).
#

from numpy import cos, pi, linspace
import matplotlib.pyplot as plt

def f(x, a=0.3, b=23, N=100):
    """
    Weierstrass' function.
    """

    return sum([a**k * cos(b**k * pi * x) for k in range(N)])


xs = linspace(-2, 2, 4001)
ys = f(xs)

plt.grid()
plt.plot(xs, ys, linewidth=0.25)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.title(r"Plot of Weierstrass' function")
plt.gcf().set_dpi(100)
plt.show()
```
