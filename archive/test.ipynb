```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 2 * np.pi, 2001)
ys = np.sin(xs)
zs = np.cos(xs)

fig, axes = plt.subplots(1, 2, figsize=(8, 5))
ax1, ax2 = axes 
ax1.plot(xs, ys)
ax2.plot(xs, zs)

```
