```python
import mpmath
mpmath.mp.dps = 100
# want to iterate (xn+2/xn)/2 until d < 10**-50 start with  x=1.4 then introduce y = f(x), 
# then calculate d = y - x and define new x -> y, 
x = mpmath.mpf('1.4')
r = 0
while True:
    y = (x+(2/x))/2
    if abs(y-x) < 10**-50:
        print(y,r,-int(mpmath.ceil(mpmath.log10(abs(y-x)))))
        break
    else:
        x = y
        r += 1
```
